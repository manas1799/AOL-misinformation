```python
import os
import json
import openai
import time
import random

# OpenAI API Key (Replace with your actual key)
OPENAI_API_KEY = "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"

# Initialize OpenAI API client
client = openai.OpenAI(api_key=OPENAI_API_KEY)

# List of JSON files to analyze
JSON_FILES = ["sleep.json", "insomnia.json", "anxiety.json"]

# Output file to store flagged posts
OUTPUT_FILE = "flagged_posts.json"

def analyze_post(post):
    """
    Uses OpenAI's model to determine if the post indicates someone suffering from sleep issues or anxiety.
    """
    prompt = f"""
    Analyze the following Reddit post and determine if the user is expressing signs of sleep deprivation or anxiety:

    Post Title: {post['title']}
    Post Content: {post.get('body', 'No additional content available')}
    Comments: {post.get('comments', 'No comments available')}

    Please respond with "YES" if the post suggests the user is suffering from sleep deprivation or anxiety.
    Otherwise, respond with "NO".
    """

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are an AI trained to detect signs of anxiety and sleep deprivation in Reddit posts."},
                {"role": "user", "content": prompt}
            ]
        )
        result = response.choices[0].message.content.strip()
        return result.lower() == "yes"

    except Exception as e:
        print(f"Error querying OpenAI: {e}")
        return False

def process_json_files():
    """
    Loads Reddit posts from JSON files, analyzes them using OpenAI, and stores relevant posts in an output file.
    """
    flagged_posts = []

    for json_file in JSON_FILES:
        if not os.path.exists(json_file):
            print(f"Skipping {json_file} (file not found).")
            continue

        print(f"Processing {json_file}...")

        with open(json_file, "r", encoding="utf-8") as file:
            for line in file:
                try:
                    post = json.loads(line)

                    # Check if the post is relevant
                    if analyze_post(post):
                        flagged_posts.append(post)

                        # Save in real-time to prevent data loss
                        with open(OUTPUT_FILE, "a", encoding="utf-8") as output:
                            output.write(json.dumps(post, ensure_ascii=False) + "\n")

                    # Delay to avoid hitting OpenAI rate limits
                    time.sleep(random.uniform(1.5, 3))

                except json.JSONDecodeError:
                    print(f"Skipping malformed entry in {json_file}.")

    print(f"Analysis complete. Flagged posts saved in {OUTPUT_FILE}.")

if __name__ == "__main__":
    process_json_files()
```
